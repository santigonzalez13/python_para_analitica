```python
# Función para limpiar y convertir la columna de tiempo
def clean_time(time_str):
    # Si el tiempo contiene paréntesis, solo toma la parte antes del espacio
    if '(' in time_str:
        return time_str.split(' ')[0]
    return time_str

# Aplica la función para limpiar la columna de tiempo
df_matches['time'] = df_matches['time'].apply(clean_time)

# Ahora, combina las columnas de 'date' y 'time' y conviértelas a datetime
df_matches['datetime'] = pd.to_datetime(df_matches['date'] + ' ' + df_matches['time'])

# Muestra la nueva columna para verificar
print(df_matches['datetime'].head())
```

```python
# Eliminar columnas innecesarias como 'Unnamed: 0' y 'notes' que no aportan información relevante
df_matches.drop(['Unnamed: 0', 'notes'], axis=1, inplace=True)

# Rellenar valores faltantes en 'attendance' con el promedio o un valor específico
df_matches['attendance'].fillna(df_matches['attendance'].mean(), inplace=True)

# Verificar y manejar valores faltantes en 'dist'
df_matches['dist'].fillna(df_matches['dist'].median(), inplace=True)
```

# Análisis descriptivo

```python
import pandas as pd

# Crear una columna para la diferencia de goles
df_matches['goal_difference'] = df_matches['gf'] - df_matches['ga']

# Calcular el promedio de goles a favor por equipo y ordenar de mayor a menor
avg_goals = df_matches.groupby('team')['goal_difference'].mean().sort_values(ascending=False)
print("Top 10 equipos por promedio de diferencia de gol:")
print(avg_goals.head(10))
```

```python
# Filtrar partidos donde la posesión fue superior al 60% y ordenar por posesión de mayor a menor
high_possession_matches = df_matches[df_matches['poss'] > 60].sort_values(by='poss', ascending=False)
print("Top 10 partidos con mayor posesión:")

high_possession_matches.head(10)


selected_columns = high_possession_matches[['ga', 'gf', 'poss', 'opponent', 'team', 'sh', 'sot']]
selected_columns
```

# Visualización de Datos con Matplotlib

Matplotlib es una biblioteca de visualización en Python que te permite crear una amplia variedad de gráficos estáticos, animados e interactivos. Vamos a comenzar con los aspectos más básicos:

Importar Matplotlib Antes de poder usar Matplotlib, necesitas importarla, generalmente se importa pyplot que es un módulo de Matplotlib:

## Crear un Gráfico Simple

Vamos a crear un gráfico de línea básico, que es uno de los gráficos más simples pero útiles en Matplotlib.

```python
import matplotlib.pyplot as plt
?plt.plot
```

```python



# Datos de ejemplo: Puntajes en un juego
juegos = [1, 2, 3, 4, 5]
puntajes = [22, 25, 27, 28, 30]

# Crear un gráfico de línea
plt.plot(juegos, puntajes)



# Mostrar el gráfico
plt.show()
```

## Agregar Títulos y Etiquetas

Para que un gráfico sea más informativo, puedes añadirle títulos y etiquetas a los ejes.

```python
# Crear un gráfico de línea
plt.plot(juegos, puntajes)

# Añadir un título
plt.title('Puntaje por Juego')

# Añadir etiquetas a los ejes
plt.xlabel('Número de Juego')
plt.ylabel('Puntaje Obtenido')

# Mostrar el gráfico
plt.show()
```

## Cambiando el Estilo y Color de la Línea

Puedes personalizar el estilo y el color de la línea para hacer el gráfico más visualmente atractivo.

```python
# Crear un gráfico de línea con línea roja y estilo de línea punteado
plt.plot(juegos, puntajes, 'r--')  # 'r--' es un formato de línea roja y punteada

# Añadir un título y etiquetas
plt.title('Puntaje por Juego')
plt.xlabel('Número de Juego')
plt.ylabel('Puntaje Obtenido')

# Mostrar el gráfico
plt.show()
```

## Guardar el Gráfico

Si deseas guardar el gráfico generado en un archivo, puedes hacerlo fácilmente.

```python
# Crear un gráfico de línea
plt.plot(juegos, puntajes, 'b-o')  # 'b-' es un formato de línea azul sólida

# Añadir un título y etiquetas
plt.title('Puntaje por Juego')
plt.xlabel('Número de Juego')
plt.ylabel('Puntaje Obtenido')

# Guardar el gráfico en un archivo PNG
plt.savefig('pruebasinformato.jpg')


# Mostrar el gráfico
plt.show()
```

## Ahora con df_matches

```python

# Histograma de goles a favor
# 'bins' especifica cuántas barras (intervalos) queremos en el histograma; aquí se establece para cada número entero de goles hasta el máximo de 'gf'.
plt.hist(df_matches['gf'], bins=range(int(df_matches['gf'].max()) + 1), color='blue', alpha=0.2)
# 'alpha' es la transparencia de las barras del histograma.
plt.title('Distribución de Goles a Favor')  # Título del gráfico
plt.xlabel('Goles')  # Etiqueta para el eje x
plt.ylabel('Número de Partidos')  # Etiqueta para el eje y
plt.show()  # Muestra el gráfico

```

Este diagrama de dispersión explora si existe una relación entre cuánto posee el balón un equipo y los goles que logra anotar en esos partidos.

```python
# Diagrama de dispersión entre la posesión del balón y los goles a favor
plt.scatter(df_matches['poss'], df_matches['gf'], alpha=0.5)  # 'alpha' controla la transparencia de los puntos
plt.title('Relación entre Posesión del Balón y Goles a Favor')
plt.xlabel('Posesión del Balón (%)')  # Etiqueta del eje x
plt.ylabel('Goles a Favor')  # Etiqueta del eje y
plt.grid(True)  # Añade una cuadrícula al gráfico para mejor visualización
plt.show()

```

Este gráfico de barras muestra el promedio de goles a favor por equipo, ordenados de mayor a menor.

```python
# Calcular el promedio de goles por equipo y ordenarlos
avg_goals_by_team = df_matches.groupby('team')['gf'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 8))  # Configura el tamaño del gráfico
avg_goals_by_team.plot(kind='bar', color='skyblue')  # Crea un gráfico de barras
plt.title('Promedio de Goles por Equipo')
plt.xlabel('Equipo')
plt.ylabel('Promedio de Goles a Favor')
plt.xticks(rotation=90)  # Rota los etiquetas de los equipos en el eje x para mejor legibilidad
plt.show()
```
