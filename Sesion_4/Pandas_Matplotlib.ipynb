```python
# Funci√≥n para limpiar y convertir la columna de tiempo
def clean_time(time_str):
    # Si el tiempo contiene par√©ntesis, solo toma la parte antes del espacio
    if '(' in time_str:
        return time_str.split(' ')[0]
    return time_str

# Aplica la funci√≥n para limpiar la columna de tiempo
df_matches['time'] = df_matches['time'].apply(clean_time)

# Ahora, combina las columnas de 'date' y 'time' y convi√©rtelas a datetime
df_matches['datetime'] = pd.to_datetime(df_matches['date'] + ' ' + df_matches['time'])

# Muestra la nueva columna para verificar
print(df_matches['datetime'].head())
```

```python
# Eliminar columnas innecesarias como 'Unnamed: 0' y 'notes' que no aportan informaci√≥n relevante
df_matches.drop(['Unnamed: 0', 'notes'], axis=1, inplace=True)

# Rellenar valores faltantes en 'attendance' con el promedio o un valor espec√≠fico
df_matches['attendance'].fillna(df_matches['attendance'].mean(), inplace=True)

# Verificar y manejar valores faltantes en 'dist'
df_matches['dist'].fillna(df_matches['dist'].median(), inplace=True)
```

# An√°lisis descriptivo

```python
import pandas as pd

# Crear una columna para la diferencia de goles
df_matches['goal_difference'] = df_matches['gf'] - df_matches['ga']

# Calcular el promedio de goles a favor por equipo y ordenar de mayor a menor
avg_goals = df_matches.groupby('team')['goal_difference'].mean().sort_values(ascending=False)
print("Top 10 equipos por promedio de diferencia de gol:")
print(avg_goals.head(10))
```

```python
# Filtrar partidos donde la posesi√≥n fue superior al 60% y ordenar por posesi√≥n de mayor a menor
high_possession_matches = df_matches[df_matches['poss'] > 60].sort_values(by='poss', ascending=False)
print("Top 10 partidos con mayor posesi√≥n:")

high_possession_matches.head(10)


selected_columns = high_possession_matches[['ga', 'gf', 'poss', 'opponent', 'team', 'sh', 'sot']]
selected_columns
```

# Visualizaci√≥n de Datos con Matplotlib

Matplotlib es una biblioteca de visualizaci√≥n en Python que te permite crear una amplia variedad de gr√°ficos est√°ticos, animados e interactivos. Vamos a comenzar con los aspectos m√°s b√°sicos:

Importar Matplotlib Antes de poder usar Matplotlib, necesitas importarla, generalmente se importa pyplot que es un m√≥dulo de Matplotlib:

## Crear un Gr√°fico Simple

Vamos a crear un gr√°fico de l√≠nea b√°sico, que es uno de los gr√°ficos m√°s simples pero √∫tiles en Matplotlib.

```python
import matplotlib.pyplot as plt
?plt.plot
```

```python



# Datos de ejemplo: Puntajes en un juego
juegos = [1, 2, 3, 4, 5]
puntajes = [22, 25, 27, 28, 30]

# Crear un gr√°fico de l√≠nea
plt.plot(juegos, puntajes)



# Mostrar el gr√°fico
plt.show()
```

## Agregar T√≠tulos y Etiquetas

Para que un gr√°fico sea m√°s informativo, puedes a√±adirle t√≠tulos y etiquetas a los ejes.

```python
# Crear un gr√°fico de l√≠nea
plt.plot(juegos, puntajes)

# A√±adir un t√≠tulo
plt.title('Puntaje por Juego')

# A√±adir etiquetas a los ejes
plt.xlabel('N√∫mero de Juego')
plt.ylabel('Puntaje Obtenido')

# Mostrar el gr√°fico
plt.show()
```

## Cambiando el Estilo y Color de la L√≠nea

Puedes personalizar el estilo y el color de la l√≠nea para hacer el gr√°fico m√°s visualmente atractivo.

```python
# Crear un gr√°fico de l√≠nea con l√≠nea roja y estilo de l√≠nea punteado
plt.plot(juegos, puntajes, 'r--')  # 'r--' es un formato de l√≠nea roja y punteada

# A√±adir un t√≠tulo y etiquetas
plt.title('Puntaje por Juego')
plt.xlabel('N√∫mero de Juego')
plt.ylabel('Puntaje Obtenido')

# Mostrar el gr√°fico
plt.show()
```

## Guardar el Gr√°fico

Si deseas guardar el gr√°fico generado en un archivo, puedes hacerlo f√°cilmente.

```python
# Crear un gr√°fico de l√≠nea
plt.plot(juegos, puntajes, 'b-o')  # 'b-' es un formato de l√≠nea azul s√≥lida

# A√±adir un t√≠tulo y etiquetas
plt.title('Puntaje por Juego')
plt.xlabel('N√∫mero de Juego')
plt.ylabel('Puntaje Obtenido')

# Guardar el gr√°fico en un archivo PNG
plt.savefig('pruebasinformato.jpg')


# Mostrar el gr√°fico
plt.show()
```

## Ahora con df_matches

```python

# Histograma de goles a favor
# 'bins' especifica cu√°ntas barras (intervalos) queremos en el histograma; aqu√≠ se establece para cada n√∫mero entero de goles hasta el m√°ximo de 'gf'.
plt.hist(df_matches['gf'], bins=range(int(df_matches['gf'].max()) + 1), color='blue', alpha=0.2)
# 'alpha' es la transparencia de las barras del histograma.
plt.title('Distribuci√≥n de Goles a Favor')  # T√≠tulo del gr√°fico
plt.xlabel('Goles')  # Etiqueta para el eje x
plt.ylabel('N√∫mero de Partidos')  # Etiqueta para el eje y
plt.show()  # Muestra el gr√°fico

```

Este diagrama de dispersi√≥n explora si existe una relaci√≥n entre cu√°nto posee el bal√≥n un equipo y los goles que logra anotar en esos partidos.

```python
# Diagrama de dispersi√≥n entre la posesi√≥n del bal√≥n y los goles a favor
plt.scatter(df_matches['poss'], df_matches['gf'], alpha=0.5)  # 'alpha' controla la transparencia de los puntos
plt.title('Relaci√≥n entre Posesi√≥n del Bal√≥n y Goles a Favor')
plt.xlabel('Posesi√≥n del Bal√≥n (%)')  # Etiqueta del eje x
plt.ylabel('Goles a Favor')  # Etiqueta del eje y
plt.grid(True)  # A√±ade una cuadr√≠cula al gr√°fico para mejor visualizaci√≥n
plt.show()

```

Este gr√°fico de barras muestra el promedio de goles a favor por equipo, ordenados de mayor a menor.

```python
# Calcular el promedio de goles por equipo y ordenarlos
avg_goals_by_team = df_matches.groupby('team')['gf'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 8))  # Configura el tama√±o del gr√°fico
avg_goals_by_team.plot(kind='bar', color='skyblue')  # Crea un gr√°fico de barras
plt.title('Promedio de Goles por Equipo')
plt.xlabel('Equipo')
plt.ylabel('Promedio de Goles a Favor')
plt.xticks(rotation=90)  # Rota los etiquetas de los equipos en el eje x para mejor legibilidad
plt.show()
```
